#05-1 함수만들기


가변 매개변수를 포함한 함수


```python
def print_n_times(n, *values):

  for i in range(n):
    print(values)
  print()

print_n_times(3, "안녕하세요", "즐거운", "프로그래밍")
```

# 05-2 함수의 활용

재귀함수


```python
# 방법1: 반복문으로 팩토리얼 구하기

def factorial(n):
  output = 1

  for i in range(1, n+1):
    output *= i

  return output

factorial(5)
```

```python
# 방법2: 재귀함수로 팩토리얼 구하기

def factorial(n):
  output = 1
  
  if n == 0:
    return 1
  
  else:
    return n * factorial(n-1)

factorial(5)
```

재귀함수의 문제
*   상황에 따라서 같은 것을 기하급수적으로 반복한다는 문제.




```python
# 재귀함수를 활용한 피보나치 수열

def fibonacci(n):
  if n == 1:
    return 1
  
  if n== 2:
    return 2
  
  else:
    return fibonacci(n-1) + fibonacci(n-2)

fibonacci(3)
```

메모화: 현재 재귀함수가 똑같은 값을 구하는 연산을 반복하기에 발생하는 문제점을 해결


```python
dictionary = {
    1 : 1,
    2: 1
}
```

```python
def fibonacci(n):
  if n in dictionary:

    return dictionary[n]

  else:
    output = fibonacci(n-1) + fibonacci(n-2)
    dictionary[n] = output # 계산한 값을 메모화하기

    return output

fibonacci(10)
```

람다

```python
# map() 함수

def power(item):
  return item * item

def under_3(item):
  return item < 3

list_input_a = [1, 2, 3, 4, 5]

# map() 함수를 사용
output_a = map(power, list_input_a)
print("# map()함수의 실행결과")
print("map(power, list_input_a):", output_a)
print("map(power, list_input_a):", list(output_a))
print()

# filter() 함수를 사용
output_b = filter(under_3, list_input_a)
print("# filter() 함수의 실행결과")
print("filter(under_3, list_input_a):", output_b)
print("filter(under_3, list_input_a):", list(output_b))
```

```python
# 람다 함수 사용 / 함수명 = lambda 매개변수: 리턴값 
power = lambda x: x * x
under_3 = lambda x: x < 3

list_input_a = [1, 2, 3, 4, 5]

# map() 함수 사용
output_a = map(power, list_input_a)
print(list(output_a))

# filter() 함수 사용
output_a = filter(under_3, list_input_a)
print(list(output_a))
```

텍스트 한 줄씩 읽기 (CSV, XML, JSON)

```python
# 랜덤하게 1000명의 키와 몸무게 만들기

import random

hanguls = list("가나다라마사아자차카타파하")

with open("info.txt", "w") as file:
  for i in range(1000):
    name = random.choice(hanguls) + random.choice(hanguls)
    weight = random.randrange(40, 100)
    height = random.randrange(140, 200)

    file.write("{}, {}, {}\n".format(name, weight, height))

with open("info.txt", 'r') as file:
  contents = file.read()
  print(contents)
```

```python
# 반복문으로 파일 한 줄씩 읽기
with open("info.txt", "r") as file:
  for line in file:
    (name, weight, height) = line.strip().split(", ")

    if (not name) or (not weight) or (not height):
      continue

    bmi = int(weight) / ((int(height) / 100) ** 2)
    result = ""
    if 25 <= bmi:
      result = "과체중"

    elif 18.5 <= bmi:
      result = "정상 체중"

    else:
      result = "저체중"

    print('\n'.join([
                     "이름 : {}",
                     "몸무게: {}",
                     "키: {}",
                     "BMI: {}"
                     "결과: {}"
    ]).format(name, weight, height, bmi, result))
    print()
```
